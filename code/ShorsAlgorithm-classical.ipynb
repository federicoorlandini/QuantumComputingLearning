```python
# Classical computer implementation
import math
import matplotlib.pyplot as mpl
import numpy as np
```

```python
# The number to factorize
N = 15

# The random number where to start
a = 13
```

```python
# First check
math.gcd(N, a)
```

```python
z = list(range(N))
y = [a**z0 % N for z0 in z]

mpl.plot(z, y)
mpl.xlabel('z')
mpl.ylabel(f'{a}^z (mod{N})')
mpl.show()
```

```python
# The period
r = z[y[1:].index(1) + 1]
print(r)
```

```python
if r % 2 == 0:
    x = a ** (r/2) % N
    print(f'x: {x}')
    if ((x + 1) % N != 0):
        print(math.gcd(int(x) + 1, N), math.gcd(int(x) - 1, N))
    else:
        print('(x + 1) % N is 0')
else:
    print('r is odd')
```
